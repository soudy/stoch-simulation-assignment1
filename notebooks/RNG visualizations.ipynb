```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from Function import uniform_sampler, latin_square_sampler, orthogonal_sampler

mpl.rcParams["figure.dpi"] = 200
mpl.rcParams["legend.fontsize"] = 11
mpl.rcParams["font.size"] = 11
plt.rc("text", usetex=True)
plt.rc("text.latex", preamble=r"""
 \usepackage{palatino}
 \usepackage{newpxmath}""")
plt.rc("font", family="serif")

n = 30
lows = (0, 0)
highs = (1, 1)
```

```python
from numpy.random import Generator, MT19937
```

```python
rng = Generator(MT19937(2412))

pure_random = uniform_sampler(rng, lows, highs, n)
x, y = pure_random.T

plt.figure(figsize=(4.5, 4))
plt.scatter(x, y, color="black", s=3)
plt.title("Pure Random Sampling")

plt.xticks([])
plt.yticks([])

plt.xlim(0, 1)
plt.ylim(0, 1)
plt.savefig('figures/pure_random_sampling.pdf', bbox_inches='tight')
plt.show()
```

```python
rng = Generator(MT19937(2412))
pure_random = latin_square_sampler(rng, lows, highs, n)
x, y = pure_random.T

minor_ticks_top = np.linspace(0, 1, 40)

fig, ax = plt.subplots(figsize=(4.5, 4))

ax.set_xticks(minor_ticks_top,minor=True)
ax.set_yticks(minor_ticks_top , minor=True)
ax.grid(which="minor", alpha=0.6, zorder=0)

ax.set_xticklabels([])
ax.set_yticklabels([])


ax.scatter(x, y, color="black", s=3, zorder=2)
ax.set_title("Latin Hypercube Sampling")
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
plt.savefig('figures/lhs_sampling.pdf', bbox_inches='tight')
plt.show()
```

```python
from matplotlib import ticker

rng = Generator(MT19937(888))
ortho = orthogonal_sampler(rng, lows, highs, n)
x, y = ortho.T

minor_ticks_top = np.arange(0, 37, 0.5)/36
major_ticks_top = np.arange(0, 37, 2)/36

fig, ax = plt.subplots(figsize=(4.5, 4))

ax.set_xticks(major_ticks_top)
ax.set_yticks(major_ticks_top)
ax.set_xticks(minor_ticks_top, minor=True)
ax.set_yticks(minor_ticks_top , minor=True)
ax.grid(which="major", alpha=0.8, zorder=0)
ax.grid(which="minor", alpha=0.2, zorder=0)

ax.set_xticklabels([])
ax.set_yticklabels([])

ax.scatter(x, y, color="black", s=3, zorder=2)
ax.set_title("Orthogonal Sampling")
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
plt.savefig('figures/ortho_sampling.pdf', bbox_inches='tight')
plt.show()
```
